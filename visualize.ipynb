```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 
# using loadtxt()
simple = np.loadtxt("simpleloss.csv",delimiter=",", dtype=float)
taylor = np.loadtxt("taylorloss.csv",delimiter=",", dtype=float)
# fourier = np.loadtxt("fourierloss.csv",delimiter=",", dtype=float)
# fouriertaylor = np.loadtxt("fouriertaylorloss.csv",delimiter=",", dtype=float)
fourieroverfit = np.loadtxt("fourierlossoverfit.csv",delimiter=",", dtype=float)
fouriertayloroverfit = np.loadtxt("fouriertaylorlossoverfit.csv",delimiter=",", dtype=float)
```

```python
simple = simple.T
taylor = taylor.T
fourier = fourier.T
fouriertaylor = fouriertaylor.T
fourieroverfit = fourieroverfit.T
fouriertayloroverfit = fouriertayloroverfit.T
```

```python
# plt.plot(simple[1],label="Simple")
# plt.plot(taylor[1],label="Taylor")
# plt.plot(fourier[1],label="Fourier")
# plt.plot(fouriertaylor[1],label="FourierTaylor")

plt.plot(fourieroverfit[1],label="Fourier")
plt.plot(fouriertayloroverfit[1],label="FourierTaylor")
plt.ylim(0,0.01)

plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend()
```
